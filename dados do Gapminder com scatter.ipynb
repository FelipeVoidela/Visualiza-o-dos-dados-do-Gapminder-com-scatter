```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("food_supply_kilocalories_per_person_and_day.csv")
```

```python
df.columns = df.columns.str.strip()

x = df['2018']
y = df['1961']
size = df.loc[:, '1961':'2018'].mean(axis=1)


norm_size = (size - size.min()) / (size.max() - size.min()) * 2000  

color = df.loc[:, '1961':'2018'].std(axis=1)  

plt.figure(figsize=(12, 8))
scatter = plt.scatter(x, y, s=norm_size, c=color, cmap='viridis', alpha=0.6, edgecolors="w", linewidth=0.5)
plt.colorbar(scatter, label='Desvio padrÃ£o do consumo de calorias')


plt.title('Consumo de Calorias por Pessoa e por Dia (1961 vs 2018)')
plt.xlabel('Consumo em 2018')
plt.ylabel('Consumo em 1961')


for i, country in enumerate(df['country']):
    plt.annotate(country, (x[i], y[i]), fontsize=9, alpha=0.75)


plt.show()
```
